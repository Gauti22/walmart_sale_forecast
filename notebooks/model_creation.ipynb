```python
import pandas as pd
import numpy as np
import pickle
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from xgboost import XGBRegressor

```

```python
data=pd.read_csv('/Users/gautammehta/Desktop/walmart_sales_forecast_project/data/post_pro/processed_data_train.csv')
data1=pd.read_csv('/Users/gautammehta/Desktop/walmart_sales_forecast_project/data/post_pro/processed_data_test.csv')
```

```python
X_train=data.iloc[:,:7]
y_train=data.iloc[:,7:]
X_test=data1.iloc[:,:7]
y_test=data1.iloc[:,7:]
```

```python
def evaluate_model(X_train,y_train,X_test,y_test,models):
        report={}
        for i in range(len(models)):
            model=list(models.values())[i]
            model.fit(X_train,y_train)

            y_test_pred=model.predict(X_test)

            test_model_score=r2_score(y_test,y_test_pred)

            report[list(models.keys())[i]]=test_model_score

        return(report)
```

```python
models={
    'LinearRegression':LinearRegression(),
    'Ridge':Ridge(),
    'KNeighborsRegressor':KNeighborsRegressor(),
    'RandomForestRegressor':RandomForestRegressor(),
    'DecisionTreeRegressor':DecisionTreeRegressor(),
    'XGBRegressor':XGBRegressor()}


model_report:dict=evaluate_model(X_train,y_train,X_test,y_test,models)
print(model_report)



best_model_score=max(sorted(model_report.values()))

best_model_name = list(model_report.keys())[
    list(model_report.values()).index(best_model_score)
]

best_model = models[best_model_name]

print(f'Best Model Found , Model Name : {best_model_name} , R2 Score : {best_model_score}')
print('\n====================================================================================\n')




```

```python
params={'n_estimators':[100,120,140],
        'max_depth':[4,6,8],
        'criterion': ['squared_error', 'absolute_error']
}
l=GridSearchCV(estimator=RandomForestRegressor(),param_grid=params,verbose=2,cv=2)
l.fit(X_train,y_train)
```

```python
o=RandomForestRegressor(criterion='squared_error',n_estimators=100,max_depth=8)
o.fit(X_train,y_train)
t_pred=o.predict(X_test)
r2_score(y_test,t_pred)
```

```python
train_r2 = o.score(X_train, y_train)   # R² on training
test_r2  = o.score(X_test, y_test)     # R² on test

print("Train R²:", train_r2)
print("Test R²:", test_r2)
```
